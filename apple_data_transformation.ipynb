```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
df = pd.read_csv("Resources/applemobilitytrends-2020-05-28.csv")
df.head()
```

```python
df.columns[6:]
```

```python
list = []
for col in df.columns[6:]:
    df_transposed=df[['geo_type', 'region', 'transportation_type', 'alternative_name', 'sub-region', 'country', col]]
    dt = []
    for i in range(0, len(df_transposed)):
        dt.append(col)
    
    df_transposed['date']=dt
    list.append(df_transposed.rename(columns={col:'score'}))
```

```python
apple_df = pd.concat(list)
apple_countries_df = apple_df.loc[df['geo_type'] == 'country/region']
apple_countries_df = apple_countries_df.loc[df['transportation_type'] == 'walking']
apple_countries_df = apple_countries_df.rename(columns={'score': 'walking_score'})
apple_countries_df.head()
```

```python
apple_countries_df.to_csv("Cleaned/apple_country_mobility_data.csv")
```
