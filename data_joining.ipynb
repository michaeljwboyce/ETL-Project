```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
apple_df = pd.read_csv("Cleaned/apple_country_mobility_data.csv")
google_df = pd.read_csv("Cleaned/google_country_mobility_data.csv")
```

```python

apple_df = apple_df.drop(apple_df.columns[0], axis=1)
apple_df.head()
```

```python
google_df = google_df.rename(columns={'country_region': 'region'})
google_df = google_df.drop(google_df.columns[0], axis=1)
google_df.head()
```

```python
apple_google_df = pd.merge(apple_df, google_df, how='left', on=['date', 'region'])
apple_google_df.drop(['country', 'sub_region_1', 'sub_region_2'], axis=1, inplace=True)
apple_google_df
```
